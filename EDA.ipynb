# **Predictive Analysis of Cardiovascular Disease Risk Using Machine Learning.**

# **Section 1: Loading Files**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style for plots
sns.set_theme(style="whitegrid")
```

# **Section 2: Loading Dataset File**

```python
# Ensure the CSV file is in the same directory as the script
df = pd.read_csv('/content/framingham.csv')
```

# **Section 3: Exploratory Data Analysis**

```python
df
```

## **Check for Missing Values**

```python
df.info()
```

```python
print("--- Missing Values Check ---")
print(df.isnull().sum())
print("\n")
```

## **Districptive Statistical Analysis**

```python
print("--- Descriptive Statistics ---")
print(df.describe())
print("\n")
```

## **Distribution of Target Variable**

```python
plt.figure(figsize=(6, 4))
# Using 'TARGET ' because of the trailing space in the original CSV column name
sns.countplot(data=df, x='TenYearCHD', palette='viridis')
plt.title('Distribution of Cardiovascular Disease Risk')
plt.xlabel('Target (0: No Risk, 1: Risk)')
plt.ylabel('Count')
plt.savefig('target_distribution.png')
plt.show()
```

## **Correlational HeatMap**

```python
plt.figure(figsize=(16, 12))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.savefig('correlation_heatmap.png')
plt.show()
```

## **Age Distribution by Risk Category**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='TenYearCHD', kde=True, bins=20, palette='magma')
plt.title('Age Distribution by Risk Category')
plt.savefig('age_distribution.png')
plt.show()
```

## **Box Plot**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='TenYearCHD', y='education', palette='Set2')
plt.title('Education vs TenYearCHD')
plt.xlabel('Target (0: No Risk, 1: Risk)')
plt.ylabel('Education')
plt.savefig('chol_boxplot.png')
plt.show()
```

## **Scatter Plot**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Updated code with corrected labels and title
sns.scatterplot(data=df, x='cigsPerDay', y='sysBP', hue='TenYearCHD', alpha=0.7, palette='coolwarm')

plt.title('Cigarettes Per Day vs. Systolic Blood Pressure')
plt.xlabel('Cigarettes Per Day')
plt.ylabel('Systolic Blood Pressure (sysBP)')

plt.savefig('bp_scatter.png')
```
